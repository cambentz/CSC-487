#### Question 3

Apply the Apriori procedure by using join operations as described on slide #15. You need to report all frequent k-itemsets. (min support = 2)

##### Step 1: Frequent 1-Itemsets (L1)

Count each item’s support:

- A -> 6
- B -> 7
- C -> 5
- D -> 2
- E -> 2

All appear ≥ 2 times -> included in L1.

**L1 = {A}, {B}, {C}, {D}, {E}**

---

##### Step 2: Generate Candidate 2-Itemsets (C2) from L1

**Join Step (self-join):** Combine all pairs from L1:

- {A, B}, {A, C}, {A, D}, {A, E}
- {B, C}, {B, D}, {B, E}
- {C, D}, {C, E}
- {D, E}

**Support Count:**

- {A, B} -> 4
- {A, C} -> 4
- {A, D} -> 1 drop
- {A, E} -> 2
- {B, C} -> 4
- {B, D} -> 2
- {B, E} -> 2
- {C, D} -> 0 drop
- {C, E} -> 2
- {D, E} -> 0 drop

**Prune Step:** Drop itemsets with support < 2.

**L2 = {A, B}, {A, C}, {A, E}, {B, C}, {B, D}, {B, E}, {C, E}**

---

##### Step 3: Generate Candidate 3-Itemsets (C3) from L2

**Join Step (L2 self-join):** Join itemsets with 2 matching items:

- Join {A, B} & {A, C} -> {A, B, C}
- Join {A, B} & {A, E} -> {A, B, E}
- Join {B, C} & {B, E} -> {B, C, E}

**Prune Step:** Keep only itemsets where all 2-subsets are in L2.

All pass the subset test.

**Support Count:**

- {A, B, C} -> 2
- {A, B, E} -> 2
- {B, C, E} -> 2

**L3 = {A, B, C}, {A, B, E}, {B, C, E}**

---

##### Step 4: Generate Candidate 4-Itemsets (C4) from L3

**Join Step:** Join {A, B, C} and {A, B, E} -> {A, B, C, E}

**Prune Step:** All 3-subsets of {A, B, C, E} are frequent -> keep 

**Support Count:**

- {A, B, C, E} -> 2

**L4 = {A, B, C, E}**

---

##### Final Result: All Frequent k-Itemsets
